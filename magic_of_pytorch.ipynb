## 1. scatter_()

to scatter (or update) elements of a tensor with values from another tensor based on specified indices. 

- scatter() vs scatter_() ; _() are in-place operations, they change the tensor directly, while scatter() is out-of-operaiton, it returns a new tensor with the scattered values.

```python
# a tensor to modify 

x = torch.zeros(2,5, dtype=torch.int64)
x
```

```python
# define indices to scatter values

idx = torch.tensor([[0], [3]])
idx
```

```python
# some values 
# exactly in the shape of x having some random values 

values = torch.randint(x.shape[0]*x.shape[1], (x.shape))
values
```

```python
x.scatter_(1, idx, values)
```
